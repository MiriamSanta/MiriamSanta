```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
media_teoretica = 6
dev_std = 4
```

```python
print("Media teoretică a distribuției:", media_teoretica)
```

```python
dimensiuni = [5, 10, 20, 50, 100, 200, 500, 1000, 5000, 10000]
medii_aleatoare = []
```

```python
for dim in dimensiuni:
    sample = np.random.normal(loc=media_teoretica, scale=dev_std, size=dim)
    medie_sample = np.mean(sample)
    medii_aleatoare.append(medie_sample)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(dimensiuni, medii_aleatoare, marker='o', label='Medii eșantioane')
plt.axhline(media_teoretica, color='red', linestyle='--', label='Media teoretică')
plt.xscale('log')
plt.xlabel('Dimensiunea eșantionului (log scale)')
plt.ylabel('Media eșantionului')
plt.title('Convergența mediei spre media teoretică')
plt.legend()
plt.grid(True)
plt.show()
```
