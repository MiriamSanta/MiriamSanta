```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
lognorm_data = np.random.lognormal(mean=0, sigma=1, size=10000)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(lognorm_data, bins=50, color='skyblue', edgecolor='black')
plt.title('Histograma valorilor log-normale (10.000 valori)')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.grid(True)
plt.show()
```

```python
media_initiala = np.mean(lognorm_data)
dev_initiala = np.std(lognorm_data)
```

```python
print("Media întregului set:", media_initiala)
print("Deviația standard a setului:", dev_initiala)
```

```python
sample_size = 30
num_samples = 100
medii_esantioane = []
for _ in range(num_samples):
    esantion = np.random.choice(lognorm_data, size=sample_size, replace=False)
    medii_esantioane.append(np.mean(esantion))
```

```python
plt.figure(figsize=(10, 5))
plt.hist(medii_esantioane, bins=20, color='orange', edgecolor='black')
plt.title('Histograma celor 100 de medii (eșantioane de dimensiune 30)')
plt.xlabel('Media eșantionului')
plt.ylabel('Frecvență')
plt.grid(True)
plt.show()
```
